## 데이터 이상치 확인

```python
def detect_and_handle_outliers(df, columns, method='remove'):
    """
    여러 컬럼의 이상치를 탐지하고 처리하는 함수
    
    Parameters:
    df (pandas.DataFrame): 입력 데이터프레임
    columns (list): 이상치를 확인할 컬럼 리스트
    method (str): 이상치 처리 방법 ('remove' 또는 'cap')
    
    Returns:
    pandas.DataFrame: 이상치가 처리된 데이터프레임
    """
    df_clean = df.copy()
    outliers_dict = {}
    
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        print(f'columns : {col}, lower_bound : {lower_bound}, upper_bound : {upper_bound}')
        print(f'columns : {col}, lower_bound_cnt : {df[(df[col] < lower_bound)].shape[0]}, upper_bound_cnt : {df[(df[col] > upper_bound)].shape[0]}')
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outliers_dict[col] = outliers[columns]
        
        if method == 'remove':
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
        elif method == 'cap':
            df_clean.loc[df_clean[col] < lower_bound, col] = lower_bound
            df_clean.loc[df_clean[col] > upper_bound, col] = upper_bound
    
    return df_clean, outliers_dict
```

```python
def check_outlier_min_max(df, columns):    
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        print('-------------------------------------------------------')
        print(f'columns : {col}, lower_bound : {lower_bound}, upper_bound : {upper_bound}')
        print(f'columns : {col}, lower_bound_cnt : {df[(df[col] < lower_bound)].shape[0]}, upper_bound_cnt : {df[(df[col] > upper_bound)].shape[0]}')
        print(f'columns : {col}, min : {df[col].min()}, median : {df[col].median()}, max : {df[col].max()}')
```
