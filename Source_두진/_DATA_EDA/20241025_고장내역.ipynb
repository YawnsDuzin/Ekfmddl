```python
df_csv['등록일시_2'] = df_csv['등록일시'].str.slice(start=0, stop=10)
df_csv.head()
```

```python
df_csv['등록일시_3'] = pd.to_datetime(df_csv['등록일시_2'])
df_csv.head()
```

```python
df_csv.info()
```

```python
df_csv['년'] = df_csv['등록일시_3'].dt.year
df_csv['월'] = df_csv['등록일시_3'].dt.month
df_csv['일'] = df_csv['등록일시_3'].dt.day
df_csv.head()
```

```python
# 고장횟수 는 년도별로 점점 늘어나고있다.
# - 자전거의 수량과 사용량이 늘어나면서 고장횟수도 느는 것으로 보인다.
df_csv['년'].value_counts().reset_index()
```

```python
df_csv['년'].value_counts().plot.bar(grid=True, figsize=(5,8), fontsize=15);  # figsize, fontsize 조정
```

```python
# 월별 고장횟수는 2,1,12,3,11 이 적다. (추울때는 이용을 별로 안해서, 고장도 주는 것 같다.)
# 여름(5,6월), 가을(9,10월)이 가장 많다.
df_csv['월'].value_counts()
```
