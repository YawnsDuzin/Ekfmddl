##### 대여시간(시간)별 평균 이용시간 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (윈도우: 'Malgun Gothic', 맥: 'AppleGothic', 리눅스: 'NanumGothic')
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우의 경우

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 대여시간을 datetime 형식으로 변환하고, 시간만 추출 (시간대별로 묶기 위해)
bike_rent['대여시간'] = pd.to_datetime(pd.to_datetime(bike_rent['대여일자'].astype(str) + ' ' + bike_rent['대여시간'].astype(str))).dt.hour

# 시간대별로 평균 이용시간 계산
avg_usage_time = bike_rent.groupby('대여시간')['이용시간(분)'].mean()

# 그래프 그리기
plt.figure(figsize=(10, 6))
avg_usage_time.plot(kind='bar')
plt.title('대여시간별 평균 이용시간')
plt.xlabel('대여시간 (시)')
plt.ylabel('평균 이용시간 (분)')
plt.xticks(rotation=0)
plt.show()
```

```python
from collections import defaultdict
file_path = r'G:\내 드라이브\DataSet\_파킷 파일\서울시 공공자전거 대여소 정보\공공자전거 대여소 정보_통합본.parquet'
station_info = em.Read_File(file_path)

districts = list(station_info['자치구'].unique())
data_by_district = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

for year in ['2018', '2019']:#['2020', '2021', '2022', '2023', '2024']:
    bike_rent = em.Read_File(r'C:\Users\user\Desktop\나따릉이\대여이력\서울특별시 공공자전거 대여이력 정보_' + str(year) + '_dropna.parquet')
    bike_rent['대여시간차'] = pd.to_datetime(bike_rent['반납일자'] + ' ' + bike_rent['반납시간'].astype(str)) - pd.to_datetime(bike_rent['대여일자'] + ' ' + bike_rent['대여시간'].astype(str))
    bike_rent['이용시간-대여시간차'] = abs(bike_rent['대여시간차'].dt.total_seconds() / 60 - bike_rent['이용시간(분)'])
    bike_rent = bike_rent[bike_rent['이용시간-대여시간차'] < 5]

    # 대여자치구 inner join
    bike_rent = pd.merge(bike_rent, station_info[['대여소번호', '자치구']], how='inner', left_on='대여대여소번호', right_on='대여소번호')
    bike_rent.drop(columns=['대여소번호'], inplace=True)
    bike_rent.rename(columns={'자치구': '대여자치구'}, inplace=True)

    # 반납자치구 inner join
    bike_rent = pd.merge(bike_rent, station_info[['대여소번호', '자치구']], how='inner', left_on='반납대여소번호', right_on='대여소번호')
    bike_rent.drop(columns=['대여소번호'], inplace=True)
    bike_rent.rename(columns={'자치구': '반납자치구'}, inplace=True)

    bike_rent.to_parquet(r'C:\Users\user\Desktop\나따릉이\대여이력\서울특별시 공공자전거 대여이력 정보_' + str(year) + '_dropna_자치구추가.parquet')

    # bike_rent = em.Read_File(r'C:\Users\user\Desktop\나따릉이\대여이력\서울특별시 공공자전거 대여이력 정보_' + str(year) + '_dropna_자치구추가.parquet')

    bike_use_time_dict = dict(bike_rent[bike_rent['대여자치구'] == bike_rent['반납자치구']].groupby('대여자치구')['이용시간(분)'].sum())
    for district, bike_use_time in bike_use_time_dict.items():
        data_by_district[district]['이용시간(분)'][year] += bike_use_time
    bike_use_dist_dict = dict(bike_rent[bike_rent['대여자치구'] == bike_rent['반납자치구']].groupby('대여자치구')['이용거리(M)'].sum())
    for district, bike_use_dist in bike_use_dist_dict.items():
        data_by_district[district]['이용거리(M)'][year] += bike_use_dist
    bike_use_cnt_dict = dict(bike_rent['대여자치구'].value_counts())
    for district, bike_use_cnt in bike_use_cnt_dict.items():
        data_by_district[district]['이용량'][year] += bike_use_cnt
```

```python
def convert_to_dataframe(data):
    rows = []
    # 자치구와 연도별 데이터에 접근
    for district, district_data in data.items():
        # 각 연도의 데이터를 추출
        for year in district_data['이용시간(분)'].keys():
            row = {
                '자치구': district,
                '연도': year,
                '이용시간(분)': district_data['이용시간(분)'][year],
                '이용거리(M)': district_data['이용거리(M)'][year],
                '이용량': district_data['이용량'][year]
            }
            rows.append(row)
    # 리스트로 정리된 데이터를 DataFrame으로 변환
    return pd.DataFrame(rows)
df = convert_to_dataframe(data_by_district)
display(df)
# df.to_parquet(r'C:\Users\user\Desktop\나따릉이\대여이력\자치구별 연도별 이용시간 이용거리 이용량_2018_2019.parquet')
```

```python
import os
folder_path = r'C:\Users\user\Desktop\나따릉이\대여이력\2019/'
# def concat_files(folder_path):
#파일 경로명 변경
file_list = os.listdir(folder_path)
file_lists = [file for file in file_list if file.endswith((".parquet"))]
columns = [
    '자전거번호', '대여일시', '대여대여소번호', '대여대여소명', '대여거치대', 
    '반납일시', '반납대여소번호', '반납대여소명', '반납거치대', '이용시간(분)', '이용거리(M)'
]
excel = pd.DataFrame(columns=columns)
file_lists.sort()

for _, file in enumerate(file_lists):
    print('path : ', folder_path + file)
    df = em.Read_File(folder_path + file)
    # display(df)
    # print(df.columns)
    # print(excel.columns)
    # display(df.info())

    # df의 컬럼명을 첫 번째 행으로 추가
    df_with_columns_as_data = pd.DataFrame([df.columns], columns=excel.columns)
    df_with_values = df.rename(columns=dict(zip(df.columns, excel)))

    # df_with_columns_as_data와 df의 값을 결합
    df_combined = pd.concat([df_with_columns_as_data, df_with_values], ignore_index=True)

    # excel과 df_combined 결합
    excel = pd.concat([excel, df_combined], ignore_index=True)
# excel.to_parquet(folder_path + 'merged.parquet')
# concat_files(folder_path)
excel
```
