---
### 출근_시, 퇴근_시 컬럼 추가

```python
df_seoul_road['출근_시'] = df_seoul_road['~08시'] + df_seoul_road['~09시']    # 출근(07 ~ 09)
df_seoul_road['퇴근_시'] = df_seoul_road['~18시'] + df_seoul_road['~19시']    # 퇴근(17 ~ 19)
```

```python
_columns = ['일자', '요일', '권역구분', '출근_시', '퇴근_시']
df_seoul_road_2 = df_seoul_road[_columns]
```

```python
df_seoul_road_2.head(2)
```

---
### 평일(월 ~ 금) 데이터만 가져오기

```python
df_seoul_road_3 = df_seoul_road_2[df_seoul_road_2['요일'].isin(['월', '화', '수', '목', '금'])]
```

```python
print(f'df_seoul_road_2.shape[0] : {df_seoul_road_2.shape[0]}')
print(f'df_seoul_road_3.shape[0] : {df_seoul_road_3.shape[0]}')
```

```python
df_seoul_road_3.head(2)
```

```python
df_seoul_road_4 = df_seoul_road_3.groupby(['권역구분']).agg({'출근_시': 'mean', '퇴근_시': 'mean'}).reset_index()
```

```python
df_seoul_road_4.shape[0]
```

```python
df_seoul_road_4.head()
```

```python
# have_bikes = rental.groupby('자치구').size()
df_seoul_road_출근 = df_seoul_road_4.groupby('권역구분')['출근_시'].mean()
df_seoul_road_퇴근 = df_seoul_road_4.groupby('권역구분')['퇴근_시'].mean()
```

### 지도에 자치구 이름 표시관련 - 위도/경도 정리

```python
dic_gus = {
    '송파구': [37.5046, 127.1187],
    '강서구': [37.5642, 126.8255],
    '강남구': [37.4915, 127.0624],
    '영등포구': [37.5244, 126.9031],
    '노원구': [37.6518, 127.0737],
    '서초구': [37.4841, 127.0126],
    '마포구': [37.5568, 126.9082],
    '강동구': [37.5527, 127.1448],
    '구로구': [37.4967, 126.8454],
    '양천구': [37.5214, 126.8595],
    '종로구': [37.5949, 126.9676],
    '중랑구': [37.5971, 127.0912],
    '은평구': [37.6205, 126.9257],
    '중구': [37.5601, 126.9985],
    '성동구': [37.5511, 127.0366],
    '동대문구': [37.5840, 127.0514],
    '용산구': [37.5348, 126.9793],
    '광진구': [37.5484, 127.0854],
    '서대문구': [37.5810, 126.9378],
    '성북구': [37.6044, 127.0147],
    '관악구': [37.4689, 126.9470],
    '도봉구': [37.6716, 127.0322],
    '금천구': [37.4640, 126.9020],
    '동작구': [37.5048, 126.9508],
    '강북구': [37.6382, 127.0116]
}
```

```python
save_name = '(folium)서울시_도로혼잡률'

# 지도 생성
# center = [37.541, 126.986]
center = [37.5665, 126.9780]

# bike_map = folium.Map(location=center, zoom_start=13, tiles='openstreetmap', name='OpenStreetMap')
bike_map = folium.plugins.DualMap(location=center, tiles="openstreetmap", name='OpenStreetMap', zoom_start=11)

# 자치구 구분 표
geo_str2 = json.load(open('G:\\내 드라이브\\DataSet\\seoul_municipalities_geo_simple.json', encoding='utf-8'))

fg0_1 = folium.FeatureGroup(name="출근", show=True).add_to(bike_map.m1)
choropleth = folium.Choropleth(geo_data = geo_str2,
                  data = df_seoul_road_출근,
                  name = '출근',
                  key_on = 'feature.properties.SIG_KOR_NM',
                  fill_color = 'Reds',
                  fill_opacity=0.7              # 채우기 불투명도
                )
choropleth.geojson.add_to(fg0_1)
# 각 구의 중심에, 자치구별 합계 표시
for idx, val in zip(df_seoul_road_출근.index, df_seoul_road_출근.values):
    folium.Marker(
        location=[dic_gus[idx][0] + 0.0055, dic_gus[idx][1]],
        popup=round(val, 2),
        icon=folium.DivIcon(html=f"<div style='font-size: 7pt; color: black; font-family: Arial; font-weight: bold;'>{round(val, 2)}</div>")
    ).add_to(fg0_1)
# =================================================================


fg0_2 = folium.FeatureGroup(name="퇴근", show=True).add_to(bike_map.m2)
choropleth = folium.Choropleth(geo_data = geo_str2,
                  data = df_seoul_road_퇴근,
                  name = '퇴근',
                  key_on = 'feature.properties.SIG_KOR_NM',
                  fill_color = 'Reds',
                  fill_opacity=0.7              # 채우기 불투명도
                )
choropleth.geojson.add_to(fg0_2)
# 각 구의 중심에, 자치구별 합계 표시
for idx, val in zip(df_seoul_road_퇴근.index, df_seoul_road_퇴근.values):
    folium.Marker(
        location=[dic_gus[idx][0] + 0.0055, dic_gus[idx][1]],
        popup=round(val, 2),
        icon=folium.DivIcon(html=f"<div style='font-size: 7pt; color: black; font-family: Arial; font-weight: bold;'>{round(val, 2)}</div>")
    ).add_to(fg0_2)
# =================================================================

# =================================================================

# FeatureGroup 생성
fg_both = folium.FeatureGroup(name="markers_both").add_to(bike_map)

# 각 구의 중심에 구 이름 표시
#fg1 = folium.FeatureGroup(name="구_이름", show=True).add_to(bike_map)
for _gu, _location in dic_gus.items():
    folium.Marker(
        location=_location,
        popup=_gu,
        name = '구_이름',
        icon=folium.DivIcon(html=f"<div style='font-size: 7pt; color: black; font-family: Arial; font-weight: bold;'>{_gu}</div>")
    ).add_to(fg_both)


# 지도 클릭 시, 좌표(위도,경로) 팝업으로 표시
folium.LatLngPopup().add_to(bike_map.m1)
folium.LatLngPopup().add_to(bike_map.m2)

#folium.TileLayer('openstreetmap', name='OpenStreetMap').add_to(bike_map)
folium.TileLayer('cartodbpositron', name='CartoDB Positron').add_to(bike_map.m1)
folium.TileLayer('cartodbdark_matter', name='CartoDB Dark Matter').add_to(bike_map.m1)
folium.TileLayer('cartodbpositron', name='CartoDB Positron').add_to(bike_map.m2)
folium.TileLayer('cartodbdark_matter', name='CartoDB Dark Matter').add_to(bike_map.m2)

folium.LayerControl().add_to(bike_map.m1)
folium.LayerControl().add_to(bike_map.m2)

# jupyter 에서 곧장 지도 표시하기
#bike_map

# web에서 표시하기
# 지도 저장
bike_map.save(save_name + ".html")
import webbrowser
webbrowser.open(save_name + ".html")
```
