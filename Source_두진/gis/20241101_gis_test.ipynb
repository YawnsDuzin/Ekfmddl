```python
g_path = 'G:\\내 드라이브\\DataSet\\gis\\'
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# 데이터 로드
df = pd.read_csv(g_path + '집계구별 직장인구.csv', encoding='cp949')
```

```python
df
```

```python
# 위도, 경도, 직장인구수 추출
data = df[['위도', '경도', '직장인구수']].values.tolist()

# 서울 중심 좌표
seoul_center = [37.5665, 126.9780]

# 지도 생성
m = folium.Map(location=seoul_center, zoom_start=11)

# 히트맵 추가
HeatMap(data).add_to(m)

# 지도 저장
m.save('seoul_workplace_density.html')
```
