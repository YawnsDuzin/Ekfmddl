### 데이터 형식 및 표시컬럼 정리
- '역번호', '역사명', '노선명', '위도', '경도'

```python
df1.rename(columns = {'역위도':'위도', '역경도':'경도'}, inplace = True)
columns = ['역번호', '역사명', '노선명', '위도', '경도']
df1_temp = df1[columns]
df1_temp.head(2)
```

```python
df2.rename(columns = {'역사_ID':'역번호', '호선':'노선명'}, inplace = True)
columns = ['역번호', '역사명', '노선명', '위도', '경도']
df2_temp = df2[columns]
df2_temp.head(2)
```

### 호선 - 확인

```python
df_temp1 = pd.DataFrame(df1_temp['노선명'].value_counts()).sort_values(by = '노선명', ascending = True)
print(df_temp1.count())
print(df_temp1)
```

```python
df_temp2 = pd.DataFrame(df2_temp['노선명'].value_counts()).sort_values(by = '노선명', ascending = True)
print(df_temp2.count())
print(df_temp2)
```

### 데이터2개 노선명 비교

```python
df_temp3 = pd.merge(df_temp1, df_temp2, how='outer',on='노선명')
df_temp3
```

### 노선명 안맞는 것들 확인?

```python
# 전체_도시철도역사정보_20240930.csv 에는 없는 노선
df_temp3[df_temp3['count_x'].isnull()]
```

```python
# 서울시 역사마스터 정보.csv 에는 없는 노선
df_temp3[df_temp3['count_y'].isnull()]
```

### 2개 데이터 역명 비교

```python
df_temp1 = pd.DataFrame(df1_temp['역사명'].value_counts()).sort_values(by = '역사명', ascending = True)
print(df_temp1.count())
print(df_temp1)
```

```python
df_temp2 = pd.DataFrame(df2_temp['역사명'].value_counts()).sort_values(by = '역사명', ascending = True)
print(df_temp2.count())
print(df_temp2)
```

```python
df_temp3 = pd.merge(df_temp1, df_temp2, how='outer',on='역사명')
df_temp3
```

```python
# 전체_도시철도역사정보_20240930.csv 에는 없는 역사명
df_temp3[df_temp3['count_x'].isnull()]
```

```python
# 서울시 역사마스터 정보.csv 에는 없는 역사명
df_temp3[df_temp3['count_y'].isnull()]
```
