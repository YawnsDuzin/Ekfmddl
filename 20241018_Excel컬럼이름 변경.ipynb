### 특정 폴더안의 Excel 파일 전체 컬럼이름 변경 후, 저장

```python
import os
import pandas as pd


# 특정 폴더의 csx, xlsx 파일들 확인
path = path_dateset + path_sub

#파일 경로명 변경
file_list = os.listdir(path)
file_lists = [file for file in file_list if file.endswith((".csv", ".xlsx"))]

print('path : ', path)

for _, file in enumerate(file_lists):
    try:
        # 파읽 읽기
        try:
            df = pd.read_csv(path + file, low_memory=False, encoding = 'cp949')
        except:
            df = pd.read_csv(path + file, low_memory=False, encoding = 'utf-8')

        # 컬럼이름 변경
        df = df.rename(columns={'기준일자' : '기준_날짜',
                                '기준_시간' : '기준_시간대', '기준시간대' : '기준_시간대',
                                '시작_대여소' : '시작_대여소_ID', '시작_대여소ID' : '시작_대여소_ID', '시작대여소ID' : '시작_대여소_ID',
                                '종료_대여소' : '종료_대여소_ID', '종료_대여소ID' : '종료_대여소_ID', '종료대여소ID' : '종료_대여소_ID',
                                '전체건수' : '전체_건수',
                                '전체이용시간(분)' : '전체_이용_분',
                                '전체이용거리(m)' : '전체_이용_거리'})

        # '집계_기준' 컬럼 없을 때 추가
        if '집계_기준' not in df.columns:
            df['집계_기준'] = None

        # 변경 된, Excel 저장
        path_save = path + 'merge/' + file
        df.to_csv(path_save, index=False, encoding='utf-8')

        print(path + file)
    except:
        print('- Err : ', file)
```
