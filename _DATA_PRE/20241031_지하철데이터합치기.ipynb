---
### 서울이 아닌 역은 삭제
 - 강일, 미사, 하남풍산, 하남시청(덕풍?신장), 하남검단산, 남위례

```python
df_train_station2[df_train_station2[' seoulmetro_code'].isnull()]
```

```python
df_train_station2 = df_train_station2[~df_train_station2[' seoulmetro_code'].isnull()]
```

```python
df_train_station2[~df_train_station2['역번호'].isnull()].head(5)
```

```python
df_train_station2['Unnamed: 5'].isnull().sum()
```

```python
df_train_station2 = df_train_station2.drop(columns=['Unnamed: 5'])
df_train_station2.head()
```

```python
df_train_station2.reset_index()
df_train_station2.head(3)
```

### 역 이름이 여러개 있는거는 일단 삭제
- 동일역에 호선이 달라서 여러개 표시되는 걸로 판단됨.

```python
# station_name(kor)으로 그룹화하고 station_code가 가장 작은 것 선택
result_df = df_train_station2.loc[df_train_station2.groupby(' station_name(kor)')['station_code'].idxmin()]
len(result_df)
```

---
### 최종 파일 저장
- 지하철역통합_20241031.parquet

```python
filename = '_최종 병합 파일\\서울교통공사 역주소 및 전화번호\\지하철역통합_20241031.parquet'
result_df.to_parquet(path_dateset + filename, index=False)
```
